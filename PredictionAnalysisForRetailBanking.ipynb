```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=0)
```

```python
#standard scaling...
# to increase accuracy
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
import keras
from keras.models import Sequential
from keras.layers import Dense
model=Sequential()
model.add(Dense(input_dim=42,init="random_uniform",output_dim=68,activation="relu")) 
model.add(Dense(init="random_uniform",output_dim=92,activation="relu")) 
model.add(Dense(init="random_uniform",output_dim=1,activation="sigmoid")) 
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
model.fit(x_train,y_train,epochs=100,batch_size=32)
```

```python
y_pred=model.predict(x_test)
y_pred
```

```python
y_pred=(y_pred>0.5)  #above 0.5 takes true
y_pred
```
